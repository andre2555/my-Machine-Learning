## Time Series Components
- Time series = a combination of:
	+ Systematic: 
		- components that have consistency or recurrence 
		- can be described and modeled
	+ Non-systematic:
		- components that cant be modeled
- Systematic components:
	+ level: avg value of the seires
	+ trend: The increasing or decreasing value in the series
	+ seasonality: The short-term repeating cycle in the series
	+ noise: random variation in the series
- Combining of Time Series Components
	+ Additive Model
		$$y(t) = level + trend + seasonality + noise$$
	+ Multiplicative Model
		$$y(t) = level * trend * seasonality * noise$$

### Additive Decomposition

```python
# Example Series
from random import randrange
series = pd.Series([i+randrange(10) for i in range(1,100)])

series.plot();
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

result = seasonal_decompose(series.values, model='additive', freq=1)
result.plot();
```

### Multiplicative Decomposition

```python
series = pd.Series([i**2.0 for i in range(1,100)])
series.plot();
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

result = seasonal_decompose(series.values, model='multiplicative', freq=1)
result.plot();
```

### Real world data Decomposition

```python
series = pd.read_csv(
    './data/airline-passengers.csv',
    header=0,
    index_col=0,
    parse_dates=True,
    squeeze=True)

series.plot();
```
