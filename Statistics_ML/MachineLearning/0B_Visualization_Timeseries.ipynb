```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Line plot

#### Dataset

```python
np.random.seed(100)
y = np.random.standard_normal(25)
x = range(len(y))

display(y)
display(x)
```

#### Simple plot

```python
# Note: x and y must match 
plt.plot(x,y)
plt.show()
```

#### Cumulative Sum of Data

```python
plt.plot(x,y.cumsum())
plt.show()
```

#### Plot with representation

```python
plt.figure(figsize=(10,5)) # width and height

plt.plot(
    x, y,
    color='b', linewidth=1.5) # line blue in color with a width of 1.5
plt.plot(
    x, y, 
    'ro', markersize=5) # red circular markers

plt.grid(True)
plt.axis ('tight')
plt.xlabel('Time')
plt.ylabel('Index')
plt.title ('Representative Cumulative plot')

plt.show()
```

#### Plot multiple dataset

```python
# Create dataset
np.random.seed(100)
y = np.random.standard_normal((25,2)).cumsum(axis=0)
x = range(len(y))
display(y[:5])

y_1 = y[:, 0]
display(y_1[:5])

y_2 = y[:, 1]
display(y_2[:5])
```

```python
plt.figure(figsize=(10,5))

plt.plot(
    x, y_1,
    color='g', lw=1.5, label = '1st dataset') # 1st dataset
plt.plot(x, y_2, 
    color='b', lw=1.5, label = '2nd dataset') # 2nd dataset
plt.plot(
    x, y,
    'rx') # x marker

plt.grid(True)
plt.legend(loc=0)
plt.axis ('tight')
plt.xlabel('Time')
plt.ylabel('Index')
plt.title ('Representative plot with two datasets')

plt.show()
```

## Histogram

```python
plt.figure(figsize=(10,5))

plt.hist(
    x=y_1,
    label='1st', bins=25)

plt.grid(True)
plt.legend(loc=0)
plt.xlabel('Index Returns')
plt.ylabel('Stock Returns')
plt.title('Histogram')

plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.hist(
    x=(y_1,y_2),
    label=['1st','2nd'], bins=25)

plt.grid(True)
plt.legend(loc=0)
plt.xlabel('Index Returns')
plt.ylabel('Stock Returns')
plt.title('Histogram')

plt.show()
```

## Scatter plot

```python
plt.figure (figsize = (7,5))
plt.scatter(
    x=y_1, y=y_2,
    marker='o')

plt.grid(True)
plt.xlabel ('1st dataset')
plt.ylabel ('2nd dataset')
plt.title('Scatter Plot')

plt.show()
```

## Plot a line with slope and intercept
- y = ax + b
    + a = slope
    + b = intercept

```python
def abline(slope, intercept):
    """Plot a line from slope and intercept"""
    axes = plt.gca()
    x_vals = np.array(axes.get_xlim())
    y_vals = intercept + slope * x_vals
    plt.plot(x_vals, y_vals, 'r--')

# Example: y = 2x + 5
abline(2,5)
```
