- Reduce Img size (Compress)
    + Down sampling faster than Conv layers
- preserve info.
- Types of pooling: Max, Min, Avg

### Pooling window size = size-reduce factor
Eg: (2x2) = reduce by half (7,7) -> (4,4)

```python
# image input
from scipy import misc

img = misc.ascent()
plt.imshow(img, cmap='gray');
print(img.shape)
```

```python
# input tensor reshape
img_tensor = img.reshape((1, 512, 512, 1))
```

### MaxPool Example

```python
from keras.layers import MaxPool2D
from keras.models import Sequential

model = Sequential()

model.add(
    MaxPool2D(
        (5, 5), # Pooling window size
        input_shape=(512, 512, 1)))

model.compile('adam', 'mse')
```

```python
img_pred = model.predict(img_tensor)[0, :, :, 0]

print(img_pred.shape)
plt.imshow(img_pred, cmap='gray');
```

### AveragePool Example

```python
from keras.layers import AvgPool2D
from keras.models import Sequential

model = Sequential()

model.add(
    AvgPool2D(
        (5, 5), # Pooling window size
        input_shape=(512, 512, 1)))

model.compile('adam', 'mse')
```

```python
img_pred = model.predict(img_tensor)[0, :, :, 0]

print(img_pred.shape)
plt.imshow(img_pred, cmap='gray');
```
