```python
from keras.layers import Embedding
from keras.models import Sequential

model = Sequential()

model.add(Embedding(input_dim=100, output_dim=2))

model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
model.summary()
```

```python
# Embedding = convert each number -> 2D vector
#      Eg: 81 -> [ 0.03724655,  0.04417845]
emb = model.predict(
    np.array([[81,  1, 96, 79],
              [17, 47, 69, 50],
              [49,  3, 12, 88]]))

emb.shape
```

# Movie Reviews Sentiment Analysis

## Dataset

```python
from keras.datasets import imdb
```

```python
# dataset set: each word is mapped -> number
(X_train, y_train), (X_test, y_test) = imdb.load_data(
    '/tmp/imdb.npz',
    num_words=None,
    skip_top=0,
    maxlen=None,
    start_char=1,
    oov_char=2,
    index_from=3)
```

```python
X_train.shape
```

```python
X_train[0]
```

### index

```python
idx = imdb.get_word_index()

idx
```

```python
# Create reverse index
rev_idx = {v+3:k for k,v in idx.items()}
rev_idx[0] = 'padding_char'
rev_idx[1] = 'start_char'
rev_idx[2] = 'oov_char'
rev_idx[3] = 'unk_char'

rev_idx
```

### Construct a review

```python
# Construct a paragraph from X_train[0]
example_review = ' '.join([rev_idx[word] for word in X_train[0]])
example_review
```

## Feature Engineerings

### Normalize the size of input

```python
print(len(X_train[0]))
print(len(X_train[1]))
print(len(X_train[2]))
```

```python
from keras.preprocessing.sequence import pad_sequences

maxlen = 100

X_train_pad = pad_sequences(X_train, maxlen=maxlen)
X_test_pad = pad_sequences(X_test, maxlen=maxlen)

print(X_train_pad.shape)
print(X_test_pad.shape)
```

```python
print(len(X_train_pad[0]))
print(len(X_train_pad[1]))
print(len(X_train_pad[2]))
```

# Model

```python
# Input
X_train_pad[0]
```

```python
# Output
y_train[0]
```

### Find the maximum number of features

```python
max_features = max([max(x) for x in X_train_pad] + 
                   [max(x) for x in X_test_pad]) + 1
max_features
```

### Model

```python
from keras.layers import LSTM, Dense
from keras.models import Sequential

model = Sequential()

model.add(Embedding(max_features, 128))
model.add(LSTM(64, dropout=0.2, recurrent_dropout=0.2))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
model.summary()
```

```python
model.fit(
    X_train_pad, y_train,
    batch_size=32, epochs=2, validation_split=0.3)
```

```python
score, acc = model.evaluate(X_test_pad, y_test)
print('Test score:', score)
print('Test accuracy:', acc)
```
