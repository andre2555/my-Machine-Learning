```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

#### Dataset

```python
np.random.seed(100)
y = np.random.standard_normal((25,2)).cumsum(axis=0)
x = range(len(y))
display(x)

y_1 = y[:, 0]
display(y_1[:5])

y_2 = y[:, 1]
display(y_2[:5])
```

# Line Plot
#### Simple plot

```python
# Note: x and y must match 
plt.plot(x,y_1)
plt.show()
```

#### Cumulative Sum of Data

```python
plt.plot(x,y_1.cumsum())
plt.show()
```

#### Plot with representation

```python
plt.figure(figsize=(10,5)) # width and height

plt.plot(
    x, y_1,
    color='b', linewidth=1.5) # line blue in color with a width of 1.5
plt.plot(
    x, y_1, 
    'ro', markersize=5) # red circular markers

plt.grid(True)
plt.axis ('tight')

plt.xlabel('Time')
plt.ylabel('Index')
plt.title ('Representative Cumulative plot')

plt.show()
```

# Line Plot multiple dataset

```python
plt.figure(figsize=(10,5))

plt.plot(
    x, y_1,
    color='g', lw=1.5, label = '1st dataset') # 1st dataset
plt.plot(x, y_2, 
    color='b', lw=2.5, label = '2nd dataset') # 2nd dataset
plt.plot(
    x, y,
    'rx') # x marker

plt.grid(True)
plt.legend(loc=0)
plt.axis ('tight')

plt.xlabel('Time')
plt.ylabel('Index')
plt.title ('Representative plot with two datasets')

plt.text(x=2, y=2, s="This is a Text", fontsize=12)

plt.show()
```

## Plot a line with slope and intercept
- y = ax + b
    + a = slope
    + b = intercept

```python
def abline(slope, intercept):
    """Plot a line from slope and intercept"""
    axes = plt.gca()
    x_vals = np.array(axes.get_xlim())
    y_vals = intercept + slope * x_vals
    plt.plot(x_vals, y_vals, 'r--')

# Example: y = 2x + 5
abline(2,5)
```

# Histogram

```python
from collections import Counter
a = np.random.randint(1,10,(50,))

cnt = Counter(a)
cnt
```

#### Plot with basic bin (interger)

```python
plt.figure(figsize=(10,5))

plt.hist(
    x=a,
    edgecolor='black')

# Choose the number of ticks
plt.xticks(np.arange(20))

plt.show()
```

#### Plot with bins

```python
plt.figure(figsize=(10,5))

plt.hist(
    x=a,
    label='data', bins=5, edgecolor='black')
        # bin size = 5

# Choose the display range
plt.axis([-1,12,0,50])
    # x: -1 -> 12
    # y: 0 -> 50

plt.grid(True)
plt.legend(loc=0)

plt.xlabel('Bins')
plt.ylabel('Counts')
plt.title('Histogram')

plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.hist(
    x=a,
    label='data', bins=[0,3,9,10], edgecolor='black')
        # [0,3]
        # [3-9]
        # [9-10]

# Choose the display range
plt.axis([-1,12,0,50])
    # x: -1 -> 12
    # y: 0 -> 50

plt.grid(True)
plt.legend(loc=0)

plt.xlabel('Bins')
plt.ylabel('Counts')
plt.title('Histogram')

plt.show()
```

#### Plot multiple data

```python
plt.figure(figsize=(10,5))

plt.hist(
    x=(y_1, y_2),
    label=['1st','2nd'], edgecolor='black')

plt.grid(True)
plt.legend(loc=0)

plt.xlabel('Index Returns')
plt.ylabel('Stock Returns')
plt.title('Histogram')

plt.show()
```

# Bar plot

```python
year = [2012, 2013, 2014, 2015, 2016, 2017]
salary = [12, 13, 14, 17, 19, 20]
population = [100, 120, 180, 250, 500, 1000]
```

```python
plt.figure (figsize = (7,5))

plt.bar(
    x=year, height=population,
    width=1, align='center', edgecolor='black')

plt.xticks(
    rotation=40)

plt.show()
```

#### Bar plot Dictionary

```python
from collections import defaultdict
cnt = defaultdict(int)
cnt = {
    'one': 11,
    'two': 22,
    'three': 33
}
```

```python
# Bar
plt.bar(
    x=range(len(cnt)), height=list(cnt.values()),
    align='center')

# x-axis labels
plt.xticks(
    ticks=range(len(cnt)),
    labels=list(cnt.keys()))

plt.show()
```

# Bubble Chart

```python
year = [2012, 2013, 2014, 2015, 2016, 2017]
salary = [12, 13, 14, 17, 19, 20]
population = [100, 120, 180, 250, 500, 1000]
```

#### Simple

```python
plt.figure (figsize = (7,5))

plt.scatter(
    x=year, y=salary, s=population,
    marker='o'
)

plt.title('Bubble Chart')
plt.show()
```

#### Colorized

```python
plt.figure (figsize = (7,5))

# Create color range
color = np.arange(len(year))

# Plot
plt.scatter(
    x=year, y=salary, s=population,
    c=color, alpha=0.7, marker='^', edgecolor='black'
)

# Add texts
for i in range(len(year)):
    plt.text(
        x=year[i] + 0.1, 
        y=salary[i] + 0.1,
        s=population[i],
        fontsize=12)

plt.title('Bubble Chart')
plt.show()
```

# Pie Chart

```python
labels = ["A","B","C","D"]
sizes = [3, 4, 6, 2]
```

```python
plt.pie(
    x=sizes,
    colors=["blue", "purple", "red", "pink"], explode = [0.1, 0.2, 0, 0],
    labels=labels, autopct="%.2f%%",  startangle=100)
plt.axis("equal")
plt.show()
```
