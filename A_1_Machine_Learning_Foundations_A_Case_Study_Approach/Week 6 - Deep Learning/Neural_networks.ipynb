```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

## Q1

```python
def f1(x1, x2, x3):
    return x1 or x2 or not x3
     
def f2(x1, x2, x3):
    return x1 and x2 and not x3

def f3(x1, x2, x3):
    return x1 or (x2 and not x3)
```

```python
def set_color(bit_value):
    if bit_value == 0:
        return 'r'
    else:
        return 'b'

functions = [f1, f2, f3]
for fun in functions:    
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    for i in range(8):
        x1 = (i & 0b001)
        x2 = (i & 0b010) >> 1
        x3 = (i & 0b100) >> 2
        ax.scatter(x3, x2, x1, c=set_color(fun(x3, x2, x1)), marker='o')
```

## Q2

```python
# (x1 AND x2) OR (NOT x1 AND NOT x2)
def f(x1, x2):
    return int((x1 and x2) or (not x1 and not x2))
```

```python
inputs = [
  [0, 0],
  [0, 1],
  [1, 0],
  [1, 1]
]

for i in inputs:
    print f(i[0], i[1])
```

```python
def Threshold(num):    
    if num > 0:
        return 1
    else:        
        return 0
```

![](./img/1.jpg)

```python
z1_bias = -0.5
z1_weight = [-1, 1]

z2_bias = 1.5
z2_weight = [-1, 1]

y_bias = -0.5
y_weight = [1, 1]

for x in inputs:    
    z1 = Threshold(z1_bias + z1_weight[0] * x[0] + z1_weight[1] * x[1])
    z2 = Threshold(z2_bias + z2_weight[0] * x[0] + z2_weight[1] * x[1])
    y = Threshold(y_bias + z1 * y_weight[0] + z2 * y_weight[1])   
    print(y)
```
